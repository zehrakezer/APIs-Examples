Scraping Multiple Pages

```python

headers = {
    'accept': '*/*',
    'accept-language': 'en-GB,en;q=0.9,tr;q=0.8,tr-TR;q=0.7',   
    'referer': 'https://www.autolist.com/listings',
    'sec-ch-ua': '"Google Chrome";v="123", "Not:A-Brand";v="8", "Chromium";v="123"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
}

##we need to loop
model=[]
mileage=[]
year=[]
dealer_name=[]
price=[]

for i in range(1,6):
    params = {
        'ads': 'true',
        'include_total_price_change': 'true',
        'include_time_on_market': 'true',
        'include_relative_price_difference': 'true',
        'latitude': '41.0247',
        'limit': '20',
        'longitude': '28.9252',
        'make': 'Tesla',
        'page': str(i),      ##our variable
        'radius': '100',
        'zip': '34010',
    }

    #response variable for the get request
    response = requests.get('https://www.autolist.com/api/v2/search', params=params, headers=headers)

    #json object
    result=response.json()

    #result items
    result=result['records']

    for i in result:
        #model
        model.append(i['model_name'])

        #mileage
        mileage.append(i['mileage'])

        #year
        year.append(i['year'])

        #dealer name
        dealer_name.append(i['dealer_name'])

        #price
        price.append(i['price'])
```

```python
tesla_pages_df= pd.DataFrame({'Model':model,'Mileage':mileage,'Year':year,'Dealer Name':dealer_name, 'Price':price})
```
